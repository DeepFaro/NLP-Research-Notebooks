```python
import pandas as pd
import time
from RAKE_JPN.rake import Rake
from xlwt import Workbook
```

```python
def get_keywordlist(input_str_jp):
    
    rake = Rake()
    start_time = time.time()
    extracted_keywords = rake.get_keywords(input_str_jp,0)
    #print(time.time()-start_time)
    #print(extracted_keywords)  # get_keywords takes two params: 1. Text, 2. Max number of keywords to return, if zero then returns all keywords.
    #print(len(extracted_keywords))
    keyword_ = {}
    for key, value in extracted_keywords.items():

        if value >= 2:
            keyword_[key] = value

    return keyword_

```

### Extract keywords and save them according to ranking

```python
import numpy as np

#get all unique ranks for each sentence

all_index = []
for index, sent in enumerate(temp_data['target_text']):
    #print(sent)
    result = get_keywordlist(sent)
    #print(result)
    temp_ = list(result.values())
    #print(temp_)
    all_index  = np.concatenate((all_index, temp_))
```

```python
col_myset = set(all_index)
len_col_myset = int(max(col_myset))
print(len_col_myset)
print(col_myset)
```

```python
wb = Workbook()
sheet1 = wb.add_sheet(str(1))


for col_name in col_myset:
    sheet1.write(0, int(col_name), col_name)

sheet1.write(0, len_col_myset+1, "origin_text")
```

```python
import string
# check english words
def is_eng_words(line):
    a_to_z = list(string.ascii_lowercase)
    A_to_Z = list(string.ascii_uppercase)
    all_alphabet = a_to_z + A_to_Z

    for character in line:
        if character in all_alphabet:
            return True
    return False
```

```python

row = 1
for index, sent in enumerate(temp_data['target_text']):
    try:
        res_keyword = get_keywordlist(sent)
        keyword_str = " | ".join(res_keyword.keys())
        if len(res_keyword) > 0 and not is_eng_words(keyword_str):
            
            #group keywords for same ranking
            # store the names (the keys of the new dict) as a set (keeps elements unique)
            names = set(res_keyword.values())
            # use a list comprehension, iterating through keys and checking the values match each n
            d = {}
            for n in names:
                d[n] = [k for k in res_keyword.keys() if res_keyword[k] == n]
            #print(d)
            for key, value in d.items():
                temp = " | ".join(value)
                
                sheet1.write(row, int(key), temp)  

            sheet1.write(row, len_col_myset+1, sent)
            row +=1
        else:
            print("no keywords")
            
    except Exception as e:
        #print(index)
        #print(sent)
        print(e)        
            
wb.save('dataset_keywords_ranking_reform.xls')    
print(sent)       
```
